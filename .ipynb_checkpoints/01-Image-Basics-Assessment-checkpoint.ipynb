#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
vertices = np.array( [[100,800] , [450,200] , [750,800]] , dtype=np.int32)
vertices.shape
vertices = vertices.reshape((-1,1,2))
vertices.shape
fotocpy = foto.copy()
cv.polylines( fotocpy , [vertices] , isClosed=True, color = (0,0,255) , thickness=20)
plt.imshow(fotocpy)
```

```python
fotocpy = foto.copy()
cv.rectangle(fotocpy, (200,400),(600,750),color=(255,0,0),thickness=5 )
plt.imshow(fotocpy)
```
